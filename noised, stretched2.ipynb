```python
import matplotlib.pyplot as plt
import librosa
import librosa.display
import os
import glob
import numpy as np 
import IPython
import random
```

```python
audio = "./data/blues.00001.wav"

data, sr = librosa.load(audio, sr=22050)

start_time = 0 
end_time = 10

start_sample = sr * start_time
end_sample = sr * end_time

data = data[start_sample : end_sample]
```

```python
# original

stft = librosa.stft(data)

db = librosa.amplitude_to_db(abs(stft))

plt.figure(figsize=(12, 4))
librosa.display.specshow(db, sr=sr, x_axis='time', y_axis='hz')
plt.savefig("./result/spectrogram2.png", bbox_inches='tight', pad_inches=0)

# add noise

noise = 0.005 * np.random.randn(*data.shape)
data_noise = data + noise

noise_stft = librosa.stft(data_noise)

noise_db = librosa.amplitude_to_db(abs(noise_stft))

plt.figure(figsize=(12, 4))
librosa.display.specshow(noise_db, sr=sr, x_axis='time', y_axis='hz')
plt.savefig("./result/spectrogram2_noise.png", bbox_inches='tight', pad_inches=0)

# stretch

rate = 0.8 + np.random.random() * 0.4
data_stretch = librosa.effects.time_stretch(data, rate=rate)

stretch_stft = librosa.stft(data_stretch)

stretch_db = librosa.amplitude_to_db(abs(stretch_stft))

plt.figure(figsize=(12, 4))
librosa.display.specshow(stretch_db, sr=sr, x_axis='time', y_axis='hz')
plt.savefig("./result/spectrogram2_stretch.png", bbox_inches='tight', pad_inches=0)
```
