```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf 
```

```python
data, sr = librosa.load("./data/blues.00001.wav")

data_stretch = librosa.effects.time_stretch(data, rate=0.8)

save_path = "./result/sample stretched.wav"

sf.write(save_path, data_stretch, sr)
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
librosa.display.waveshow(data, sr=sr, alpha=0.8)
plt.title('Original')

plt.subplot(1, 2, 2)
librosa.display.waveshow(data_stretch, sr=sr, alpha=0.8)
plt.title('Stretched')

plt.show()
```
