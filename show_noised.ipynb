```python
import numpy as np
import soundfile as sf 
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
file_path = './data/blues.00001.wav'

data, sr = sf.read(file_path)

ipd.Audio(data, rate=sr)
```

```python
noise_amp = 0.01 * np.random.uniform() * np.amax(data)

noise = noise_amp * np.random.normal(size=data.shape)

data_noise = data.astype('float32') + noise.astype('float32')

save_path = './result/sample with noise.wav'

sf.write(save_path, data_noise, sr)

sample_data, sample_sr = sf.read(save_path)

ipd.Audio(sample_data, rate=sample_sr)
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(data)
plt.title('Original')

plt.subplot(1, 2, 2)
plt.plot(sample_data)
plt.title('Noise added')

plt.show()
```
